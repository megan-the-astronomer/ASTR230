```python
# import necessary modules 
import numpy as np
from astropy.io import fits
```

```python
# open fits file
ffile = fits.open('/data/iraf/mr82/raw/hlsp_carina_hst_wfc3-uvis_14_f657n_v1_drz_sci.fits')
```

The 'ffile' object is a list of each header-data-unit (HDU) in the fits file. To see how many it found, type:

```python
ffile.info()
```

In our case there should only be a [0]. This data object is by default returned as a 32-bit floating point array. To make the data easy to access, you can assign the data array to its own variable. 

```python
fdata = ffile[0].data 
```

To check the dimensions of the array:

```python
fdata.shape
```

remember that python gives the dimensions as (y,x)! Compare these dimensions to what you see when you look at the image in ds9. 

The header contains a lot of useful information about the data specified with different key words. Take a look at the header using the ".header" extension of the data. You can also write this to a separate variable. 

```python
fheader=ffile[0].header
```

```python
fheader
```

If you know the keyword, you can print just the value of that specific header keyword. 

```python
fheader['OBJECT']
```

Header keywords can be overwritten and/or updated with new information. Try changing the name of the 'OBJECT' in the header, then print the results to see if it worked.  

```python
fheader['OBJECT'] = 'Mystic Mountains' 
```

```python
fheader
```

Finally, we can write the data and updated header to a new fits file. Note the final keyword "overwrite=True" -- if you run this notebook more than once, this will overwrite the file with the new version. If we did not include this keyword, python would complain that the file already exists and would not save the 'new' version.

```python
fits.writeto('/data/iraf/mr82/reduced/my_new_HH_image.fits', fdata, fheader, overwrite=True)
```
