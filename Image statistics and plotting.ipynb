```python
# import required modules
import numpy as np 
from matplotlib import pyplot as plt
from astropy.io import fits 
```

```python
# open fits file
ffile = fits.open('/Users/mr82/teaching/ASTR230_spring2024/lab2/lab2_data/hlsp_carina_hst_wfc3-uvis_14_f657n_v1_drz_sci.fits')
```

```python
fdata = ffile[0].data 
fheader=ffile[0].header
```

To start, let's display the data just using the defaults. 

```python
plt.imshow(fdata)
```

The default intensity scaling is a linear map between the highest and lowest values in the data. We can look at the image statistics to see if this is a good fit for these data. 

```python
fdata.min()
```

```python
fdata.max()
```

```python
np.median(fdata)
```

While there are some very high values in the array, most of the pixels have a value < 1. We can change the upper and lower limits that imshow uses to scale the data with the keywords vmin and vmax. 

```python
plt.imshow(fdata, vmin=0.125, vmax=1.25)
```

Notice the axis labels. The x-axis runs left to right as usual, but the y-axis has zero at the *top*. The image is rotated 180 degrees compared to the display in ds9. To get a more familiar orientation, you can use origin='lower'. 

```python
plt.imshow(fdata, origin='lower', vmin=0.125, vmax=1.25)
```

The default colormap for python is viridis. Curious about why? See the talk on the subject given at SciPy 2015: https://www.youtube.com/watch?v=xAoljeRJ3lU

You can change the colormap using the cmap keyword. Let's try showing the image in grayscale.

```python
plt.imshow(fdata,origin='lower',cmap='gray',vmin=0.125,vmax=1.25)
```

... and inverted grayscale

```python
plt.imshow(fdata,origin='lower',cmap='binary',vmin=0.125,vmax=1.25)
```

Much, much more about colormaps:
https://matplotlib.org/stable/users/explain/colors/colormaps.html

You can also plot subsets of images using array indexing. Remember that python coordinates are written (y,x)!

```python
plt.imshow(fdata[2750:5750,4000:6500],origin='lower',cmap='binary',vmin=0.125,vmax=1.5)
```

Finally, you can save your image to a .png or .pdf file to include in your lab write up. 

```python
plt.imshow(fdata[2750:5750,4000:6500],origin='lower',cmap='binary',vmin=0.125,vmax=1.5)
plt.savefig('myMM.png')
```
